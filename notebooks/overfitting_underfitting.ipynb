Underfitting ocurre cuando un modelo es demasiado simple para captar la estructura subyacente de los datos. 

En este caso, el modelo no logra aprender ni la señal ni el ruido, resultando en un rendimiento pobre tanto en el conjunto de entrenamiento como en el conjunto de prueba.

## Características del Underfitting:

Bajo rendimiento en el conjunto de entrenamiento: El modelo presenta alta pérdida y baja precisión incluso en los datos que ha visto.

Rendimiento similar en nuevos datos: El modelo también tiene un rendimiento deficiente en el conjunto de validación o prueba.

Simplicidad del modelo: Modelos demasiado simples (como una línea recta en un problema no lineal) tienden a caer en underfitting.

## Soluciones para el Underfitting:

💡 Incrementar la complejidad del modelo: Elegir un modelo más complejo que pueda capturar las relaciones en los datos, como un polinomio de mayor grado.

💡 Ajustar los hiperparámetros: Optimizar parámetros que controlan la capacidad del modelo, ej: profundidad de un árbol de decisión.

💡 Ajustar funciones de activación o arquitecturas: En redes neuronales, experimentar con diferentes capas o funciones de activación.

## Bias (Sesgo)
Se refiere a la suposición que hace un modelo sobre cómo funcionan los datos. 
Es el error que ocurre porque el modelo es demasiado simple para representar la complejidad de los datos reales. Un alto sesgo implica que el modelo no capta adecuadamente la estructura subyacente del problema, lo que lleva a un desempeño ineficiente, comúnmente bajo tanto en el conjunto de entrenamiento como en el conjunto de prueba.

Overfitting ocurre cuando un modelo de machine learning se ajusta demasiado bien a los datos de entrenamiento. En otras palabras, el modelo aprende no solo la señal (las relaciones verdaderas y generales) de los datos, sino también el ruido y las peculiaridades específicas de esos datos. Esto resulta en un modelo que tiene un rendimiento excepcional en el conjunto de entrenamiento, pero un rendimiento deficiente en datos no vistos (conjunto de prueba o validación) porque no ha generalizado bien.

Características del Overfitting:
Alto rendimiento en el conjunto de entrenamiento: El modelo tiene baja pérdida y alta precisión en los datos de entrenamiento.
Mal rendimiento en nuevos datos: En el conjunto de validación o prueba, el modelo presenta una alta tasa de error.
Complexidad del modelo: Modelos muy complejos (como redes neuronales profundas sin regularización) son más propensos a overfitting.
Soluciones para el Overfitting:
Regularización: Técnicas como L1 (Lasso) o L2 (Ridge) añaden un término de penalización al modelo.
Poda (Pruning): En árboles de decisión, se pueden eliminar ramas que no contribuyen significativamente al rendimiento del modelo.
Aumento de datos: Generar más datos o utilizar técnicas de data augmentation para mejorar la generalización.
Detener antes el entrenamiento: Usar early stopping para evitar el sobreajuste durante el entrenamiento en redes neuronales.

## Overrfitting

Ocurre cuando un modelo de machine learning se ajusta demasiado bien a los datos de entrenamiento. 
En otras palabras, el modelo aprende no solo la señal (las relaciones verdaderas y generales) de los datos, sino también el ruido y las peculiaridades específicas de esos datos. 

Esto resulta en un modelo que tiene un rendimiento excepcional en el conjunto de entrenamiento, pero un rendimiento deficiente en datos no vistos (conjunto de prueba o validación) porque no ha generalizado bien.

## Características del Overfitting:

Alto rendimiento en el conjunto de entrenamiento: El modelo tiene baja pérdida y alta precisión en los datos de entrenamiento.

Mal rendimiento en nuevos datos: En el conjunto de validación o prueba, el modelo presenta una alta tasa de error.

Complexidad del modelo: Modelos muy complejos (como redes neuronales profundas sin regularización) son más propensos a overfitting.

Soluciones para el Overfitting:

💡 Regularización: Técnicas como L1 (Lasso) o L2 (Ridge) añaden un término de penalización al modelo.

💡 Poda (Pruning): En árboles de decisión, se pueden eliminar ramas que no contribuyen significativamente al rendimiento del modelo.

💡 Aumento de datos: Generar más datos o utilizar técnicas de data augmentation para mejorar la generalización.

💡 Detener antes el entrenamiento: Usar early stopping para evitar el sobreajuste durante el entrenamiento en redes neuronales.

🕵️ La varianza se refiere a la sensibilidad del modelo a pequeñas fluctuaciones en los datos de entrenamiento. 
Un modelo con alta varianza capta todos los detalles y ruido en los datos de entrenamiento, lo que puede llevar a que el modelo se ajuste demasiado a esos datos específicos, y, por lo tanto, no generalice bien a datos no vistos. 

Esto significa que el modelo funcionará excepcionalmente bien en el conjunto de entrenamiento, pero mal en el conjunto de prueba.

## Posibles razones de ruido: 

🦹 Errores de medición 

🦹 Puntos de datos que se desvían significativamente de otros en el conjunto. Estos pueden ser errores de medición o datos realmente inusuales. (outliers)

🦹 En los datos de series temporales, condiciones y relaciones pueden cambiar a lo largo del tiempo.

🦹  Si los datos no incluyen todas las variables relevantes, el modelo puede ajustar su predicción a patrones que no son de hecho ciertas
